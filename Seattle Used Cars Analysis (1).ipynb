```python
# Count how many duplicates we have for every unique vehicle
dup_df = cars_df.groupby(["model_full_name", "miles"]).agg(total_occurences = ("model_full_name", "count"))

# Printing dup_df
dup_df.sort_values(["total_occurences"], ascending = False)
```

```python
# Dropping the Duplicates
cars_df = cars_df.drop_duplicates(subset=["model_full_name", "miles"])
print("There are", len(cars_df), "used vehicles available at the Seattle Dealerships")
```

# Market Analysis

### 1) Which brands of used cars are most available in Seattle?

```python
brands_df = cars_df.groupby(["brand"]).agg(total_available = ("brand", "count"))
brands_df.sort_values(["total_available"], ascending = False)
```

### 2. Which sub models of used cars are most available in Seattle?

```python
sub_model_df = cars_df.groupby(["sub_model", "brand"]).agg(total_available = ("sub_model", "count"))
sub_model_df.sort_values(["total_available"], ascending = False).head (25)
```

### 3. Which specific models of used vehicles are available in Seattle?

```python
model_df = cars_df.groupby(["model_full_name"]).agg(total_available = ("model_full_name", "count"))
model_df.sort_values(["total_available"], ascending = False)
```

```python
model_df = cars_df.groupby(["year", "brand", "sub_model"]).agg(total_available = ("sub_model", "count"))
model_df.sort_values(["total_available"], ascending = False)
```

### 4. What are the total vehicles for each value type?

```python
value_type_df = cars_df.groupby(["value_type"]).agg(total_available = ("value_type", "count"))
value_type_df.sort_values(["total_available"], ascending = False)
```

### Which vehicles should we consider for purchase?

**Criteria:** For a used car to qualify as a good purchase, it must:**

- Have Driven Less than 45,000 miles
- Be a Model in or after 2018
- Have a value type of "great Value"
- Civic, Corolla, 1500, Crosstrek, F-150, Tucson

```python
shortlist_df = cars_df.loc[(cars_df["miles"]<=45000)
                          & (cars_df["year"]>=2018)
                          & (cars_df["value_type"] == "great Value")
                          & ((cars_df["sub_model"] == "Civic") |
                            (cars_df["sub_model"] == "Corolla") |
                            (cars_df["sub_model"] == "F-150") |
                            (cars_df["sub_model"] == "Crosstrek") |
                            (cars_df["sub_model"] == "Tucson") |
                            (cars_df["sub_model"] == "1500"))] 
                            
len(shortlist_df)
```

```python
shortlist_df.sort_values(["sub_model"])
```

## Finding the Best Honda Civic in the Market

```python
shortlist_df = cars_df.loc[(cars_df["miles"]<=45000)
                          & (cars_df["year"]>=2018)
                          & (cars_df["value_type"] == "great Value")
                          & ((cars_df["sub_model"] == "Civic"))]

shortlist_df["miles_price_ratio"] = shortlist_df["miles"]/ shortlist_df["price"]

print("There are", len(shortlist_df), "used Honda Civics in the market that meet our criteria. They are:")
display(shortlist_df.sort_values(["miles_price_ratio", "miles"]))
```

## Build a Report Maker Function That Accepts a Year, Brand, and Sub Model and creates a nice report

```python
def get_report (year, brand, sub_model):
    
    # Printing the Vehicle Name
    print("Vehicle Name:", year, brand, sub_model)

    requests_df = cars_df.loc[(cars_df["year"] == year)
                             & (cars_df["brand"] == brand)
                             & (cars_df["sub_model"] == sub_model)]

    # Calculating the Total Available Cars, Median Miles of Available Cars, and Median Price of Available Cars
    
    print("Cars Available:", len(requests_df))
    print ("Median Miles:", requests_df["miles"].median())
    print("Median Price:", requests_df["price"].median())

    requests_df["miles_price_ratio"] = requests_df["miles"]/requests_df["price"]

    print("The available vehicles are:")
    display(requests_df)


    shortlist_df = requests_df.loc[(requests_df["miles"]<= 45000)
                                  & (requests_df["value_type"]=="great Value")]

    print("Our", len(shortlist_df), "recommendations are:")
    display(shortlist_df.sort_values(["miles_price_ratio"]))
    
    print("Prices to Miles Relationship Scatter Plot")
    requests_df.plot.scatter(x='miles', y='price', c='Red')
```

```python
get_report(2019, "Toyota", "Highlander")
```
